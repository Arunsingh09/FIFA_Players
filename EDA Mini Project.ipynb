## 3.	Generate a table containing the top 20 players ranked by Overall score and whose contract expires in 2020.
a)	What would the average wage for this set of players be?
b)	What is the average age?
c)	Is there a correlation between the Overall rating and Value for these players?



```python
d1[(d1['Overall']) & (d1['Contract Valid Until'] == '2020')].head(20)
```

```python
#a) What would the average wage for this set of players be?
d1[(d1['Overall']) & (d1['Contract Valid Until'] == '2020')].head(20).groupby('Wage').mean()
```

```python
#b) What is the average age?
d1[(d1['Overall']) & (d1['Contract Valid Until'] == '2020')].head(20).groupby('Age').mean()
```

```python
#c) Is there a correlation between the Overall rating and Value for these players?
d2 = d1[(d1['Overall']) & (d1['Contract Valid Until'] == '2020')].head(20)
d2[['Overall','Value']].corr()
```

## Overall rating is Highly correlated with value
